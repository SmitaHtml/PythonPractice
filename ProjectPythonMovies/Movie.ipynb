```python
#now lets find out how many animation vs comedy movies exist in this dataset

movies_df['Genre_arr']= movies_df['Genre'].str.split(',')
movies_df.head()
```

```python
movies_lambda = lambda x: set(['Action','Romance','Comedy']).issubset(x)
mix_genres_movies = movies_df[movies_df.Genre_arr.map(movies_lambda)] 
print("# of mix_genres_movies")
len(mix_genres_movies.index)
#mix_genres_movies
```

```python
# Create new column name genres_count
counter_lambda = lambda x: len(x)
movies_df['Genre_count'] = movies_df.Genre_arr.apply(counter_lambda)
movies_df.head()
```

```python
# now lets analyze how many genres do various movies have by plotting a histogram of genres

num_bins = 3
plt.hist(movies_df.Genre_count,num_bins, color='green', alpha=0.8)

plt.title('')
plt.xlabel('# of genres')
plt.ylabel('# of movies')
plt.axis([0, 9, 0, 1000])
plt.show()
```

```python
# lets see how many movies are there in each genre?
from collections import Counter
flattened_Genres = [item for sublist in movies_df.Genre_arr for item in sublist]
Genre_dict = dict(Counter(flattened_Genres))
print (Genre_dict)
```

```python
# lets plot the Genre distrinution as a pie chart
plt.pie(Genre_dict.values(), labels = Genre_dict.keys())
plt.title('Genre distribution of movies')
plt.show
```

```python
# lets plot a bar chart
x = list(range(len(Genre_dict)))
plt.xticks(x, Genre_dict.keys(), rotation =80)
plt.bar(x,Genre_dict.values())
plt.grid()
plt.plot()
```
